```python
# costumers
costumer = pd.read_excel("ليست مشتريان.xlsx")
costumer = costumer.rename(columns= {"كد گروه مشتري": "Status"})
costumer = costumer.rename(columns= {"كد": "Code"})

costumer = costumer[["Status", "Code"]]
costumer_dict = costumer.to_dict(orient="index")

# make a list of costumer ids
costumer_ids = []
for i in range(len(costumer_dict)):
    costumer = costumer_dict[i]
    ID = costumer["Code"]
    costumer_ids.append(ID)

# make a list of costumer status
costumer_status = []
for i in range(len(costumer_dict)):
    costumer = costumer_dict[i]
    status = costumer["Status"]
    costumer_status.append(status)


# creating a dict
costumer_dict = dict(zip(costumer_ids, costumer_status))
costumer_dict
```

```python
import pandas as pd
# excel
sale = pd.read_excel("مرداد تا مهر.xlsx")
sale = sale.rename(columns= {"مشتري": "Costumer"})
sale = sale.rename(columns= {"مقدار-فرعي": "Gram"})
sale = sale.rename(columns= {"مبلغ فروش": "Value"})
sale = sale.rename(columns= {"كد مشتري": "Code"})
sale["Status"] = None

# delete last row
sale = sale.iloc[:-1]
sale = sale[["Costumer", "Gram", "Value", "Code", "Status"]]


for i in range(len(sale)):
    code = sale.loc[i, "Code"]
    try:
        status = costumer_dict[code]
        if status == 1:
            sale.loc[i, "Status"] = "Person"
        elif status == 2:
            sale.loc[i, "Status"] = "Cafe/Restaurant"

        elif status == 4:
            sale.loc[i, "Status"] = "Company"
    except:
        continue
        
sale           
```

```python
# gram sum by each group
gram_sum = sale.groupby("Status")["Gram"].sum()

cafe_gram_sum = gram_sum[0]
company_gram_sum = gram_sum[1]
person_gram_sum = gram_sum[2]


percentage_cafe = (cafe_gram_sum / (cafe_gram_sum + company_gram_sum + person_gram_sum)) * 100
percentage_company = (company_gram_sum / (cafe_gram_sum + company_gram_sum + person_gram_sum)) * 100
percentage_person = (person_gram_sum / (cafe_gram_sum + company_gram_sum + person_gram_sum)) * 100


print(f"sum gram of cafe & restaurants purchases = {cafe_gram_sum //1000} kg --> {percentage_cafe}%")
print(f"sum gram of companies purchases = {company_gram_sum//1000} kg --> {percentage_company}%")
print(f"sum gram of persons purchases = {person_gram_sum//1000} kg --> {percentage_person}%")
```

```python
# value sum by each group 
value_sum = sale.groupby("Status")["Value"].sum()

cafe_value_sum = value_sum[0]
company_value_sum = value_sum[1]
person_value_sum = value_sum[2]

percentage_cafe = (cafe_value_sum / (cafe_value_sum + company_value_sum + person_value_sum)) * 100
percentage_company = (company_value_sum / (cafe_value_sum + company_value_sum + person_value_sum)) * 100
percentage_person = (person_value_sum / (cafe_value_sum + company_value_sum + person_value_sum)) * 100

print(f"sum value of cafe & restaurants purchases = {cafe_value_sum} Rial --> {percentage_cafe}%")
print(f"sum value of companies purchases = {company_value_sum} Rial --> {percentage_company}%")
print(f"sum value of persons purchases = {person_value_sum} Rial --> {percentage_person}%")


# removing reapeted costumer
set_df = sale[["Code","Status"]]
unique_ids = set_df["Code"].drop_duplicates().tolist()

```

```python
# counting each group unique numbers
person_ids = []
cafe_ids = []
company_ids = []

for i in unique_ids:
    try:
        status = costumer_dict[i]
        if status == 1:
            person_ids.append(i)

        elif status == 2:
            cafe_ids.append(i)

        elif status == 4:
            company_ids.append(i)
    except:
        continue
print( f" in this span we had  {len(person_ids)} unique Persons who purchased from XAV")
print(f" in this span we had {len(cafe_ids)} unique Cafe & Restaurants who purchased from XAV")
print(f" in this span we had {len(company_ids)} unique Companies who purchased from XAV")
```
