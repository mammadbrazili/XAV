## بهای تمام شده سرت شده

```python
# اکسل بهای تمام شده
sort = pd.read_excel("بهاي تمام شده سورت شهريور 1402.xlsx")
sort = sort.iloc[:-1]

# adding name column
sort["Name"] = None

# adding Name column
for j in range(len(sort)):
    name = sort.loc[j,"عنوان كالا"]
    list_name = []


    # delete gram part
    index = next((i for i, c in enumerate(name) if not c.isalpha() and not c.isspace()), len(name))

    # english name + persian
    english_name = name[:index].strip()
    english_part = re.sub(r'[^a-zA-Z\s]', '', english_name)

    # Check if an English part was found
    if english_part:
        list_name.append(english_part)
    
    sort.loc[j, "Name"] = english_part
    
sort = sort.rename(columns= {"ورودي-في متوسط": "Cost_Sort"})
sort = sort.rename(columns = {"كد كالا": "Code"})
sort = sort .drop ("عنوان كالا", axis=1)

# بهای تمام شده رست به ازای یک کیلو و اسم منتاظر آنها
sort_name_value_dict = {}

for i in range(len(sort)):
    name = sort.loc[i, "Name"]
    cost = sort.loc[i, "Cost_Sort"]
    sort_name_value_dict[name] = cost
    
sort
```

## بهای تمام شده ‍‍پکیج

```python
package = pd.read_excel("بهاي تمام شده شهريور 1402.xlsx")
package = package.iloc[:-1]

# adding series to a column
for index, row in package.iterrows():
    code = row["كد كالا"]
    series = str(int(code//100))[-1]
    if series == "1":
        package.at[index, 'Series'] = 'X'
        
    elif series == "3":
        package.at[index, "Series"] = "A"
        
    elif series == "5":
        package.at[index, "Series"] = "V"
        
# adding gram
for index, row in package.iterrows():
    series = row["Series"]
    code = row["كد كالا"]
    gram = str(code)[-1]
    
    if series == "A":
        if gram == "1":
            package.at[index, "Gram"] = 150
        if gram == "2":
            package.at[index, "Gram"] = 750
        if gram == "3":
            package.at[index, "Gram"] = 40
            
    elif series == "X":
        if gram == "1":
            package.at[index, "Gram"] = 80
        if gram == "2":
            package.at[index, "Gram"] = 80
        if gram == "3":
            package.at[index, "Gram"] = 40
            
    elif series == "V":
        if gram == "1":
            package.at[index, "Gram"] = 150
        if gram == "2":
            package.at[index, "Gram"] = 750
        if gram == "3":
            package.at[index, "Gram"] = 4000               
   
    

# adding Name column
for j in range(len(package)):
    name = package.loc[j,"عنوان كالا"]
    list_name = []



    # delete gram part
    index = next((i for i, c in enumerate(name) if not c.isalpha() and not c.isspace()), len(name))

    # english name + persian
    english_name = name[:index].strip()
    english_part = re.sub(r'[^a-zA-Z\s]', '', english_name)

    # Check if an English part was found
    if english_part:
        list_name.append(english_part)
    
    package.loc[j, "Name"] = english_part
    
package = package.rename(columns= {"ورودي-في متوسط": "Cost_Package"})
package = package.rename(columns = {"كد كالا": "Code"})
package = package .drop ("عنوان كالا", axis=1)
package = package[["Code", "Name", "Series","Gram", "Cost_Package" ]]

# بهای تمام شده پکیج به ازای یک کیلو و اسم منتاظر آنها
package_name_value_dict = {}

for i in range(len(package)):
    code = package.loc[i, "Code"]
    cost = package.loc[i, "Cost_Package"]
    package_name_value_dict[code] = cost
    

    
package
```

## درصد سود  

```python
# پیدا کردن ستون های مشترک
package_name_list = list(set(package["Name"]))
sort_name_list = list(set(sort["Name"]))
saled_name_list = list(set(fi_forosh["Name"]))

# keep the same Names
common_names = list(set(package_name_list).intersection(saled_name_list))

# keep the common name column in saled 
profit = fi_forosh[fi_forosh['Name'].isin(common_names)]

# اضافه کردن بهای تمام شده پکیج شده 
profit["Cost_Packaged"] = profit["Code"].map(package_name_value_dict)

# اضافه کردن بهای تمام شده رست شده--> به ازای ۱ کیلو
profit["Cost_Sorted"] = profit["Name"].map(sort_name_value_dict)
# تبدیل به وزن 
profit["Cost_Sorted"] = (profit["Cost_Sorted"] / 1000) * profit["Gram"]


# اضافه کردن بهای تمام شده گرین بین --> به ازای ۱ کیلو
profit["Cost_GreenBean"] = None
profit = profit.reset_index(drop=True)
for i in range(len(profit)):
    code = profit.loc[i, "Code"]
    code = int(str(code)[1:4])
    profit.loc[i, "Cost_GreenBean"] = gb_name_value_dict[code]
    
# تبدیل به وزن 
profit["Cost_GreenBean"] = (profit["Cost_GreenBean"]/ 1000) * profit["Gram"]

# calculating profit
profit["GB_Profit %"] =((profit["Price"] - profit["Cost_GreenBean"]) / profit["Price"])*100
profit["SB_Profit %"] =((profit["Price"] - profit["Cost_Sorted"]) / profit["Price"]) * 100
profit["P_Profit %"] =((profit["Price"] - profit["Cost_Packaged"]) / profit["Price"]) * 100

# calculating profit for 1 kg
profit["GB%"] = None
profit["SB%"] = None
profit["P%"] = None

# keeping highest gram of each product
for i in range(len(profit)):
    series = profit.loc[i, "Series"]
    code = profit.loc[i, "Code"]
    price = profit.loc[i, "Price"]
    package_cost = profit.loc[i, "Cost_Packaged"]
    sorted_cost = profit.loc[i, "Cost_Sorted"]
    gb_cost = profit.loc[i, "Cost_GreenBean"]
    
    if series == "A":
        if str(code)[-1] == "2":
            price_1k = price * 1.33
            package_cost_1k = package_cost * 1.33
            sorted_cost_1k = sorted_cost * 1.33
            gb_cost_1k = gb_cost*1.33
            
            profit.loc[i, "GB%"] = ((price_1k - gb_cost_1k)/ price_1k) *100
            profit.loc[i, "SB%"] = ((price_1k - sorted_cost_1k)/ price_1k) *100
            profit.loc[i, "P%"] = ((price_1k - package_cost_1k)/ price_1k) *100
            
    elif series == "V":
        if str(code)[-1] == "3":
            price_1k = price / 4
            package_cost_1k = package_cost / 4
            sorted_cost_1k = sorted_cost / 4
            gb_cost_1k = gb_cost / 4

            profit.loc[i, "GB%"] = ((price_1k - gb_cost_1k)/ price_1k) *100
            profit.loc[i, "SB%"] = ((price_1k - sorted_cost_1k)/ price_1k) *100
            profit.loc[i, "P%"] = ((price_1k - package_cost_1k)/ price_1k) *100

        elif str(code)[-1] == "2":
            price_1k = price * 1.33
            package_cost_1k = package_cost * 1.33
            sorted_cost_1k = sorted_cost * 1.33
            gb_cost_1k = gb_cost*1.33

            profit.loc[i, "GB%"] = ((price_1k - gb_cost_1k)/ price_1k) *100
            profit.loc[i, "SB%"] = ((price_1k - sorted_cost_1k)/ price_1k) *100
            profit.loc[i, "P%"] = ((price_1k - package_cost_1k)/ price_1k) *100
            
# debugging v series           
for i in range(len(profit)-1):
    first_row = profit.loc[i, "GB%"]
    next_row = profit.loc[i+1, "GB%"]
    
    if first_row and next_row != None:
        profit.loc[i, "GB%"] = None
        profit.loc[i, "SB%"] = None
        profit.loc[i, "P%"] = None


            




profit.to_excel('Profit.xlsx', index=False)


profit
```
