```python
import pandas as pd
df = pd.read_excel("/Users/mohammad/Dropbox/XAV Finance (1)/گزارشات/1403/فروردین/نرخ تولید 1402.xlsx",sheet_name="gb-roast")[:-2].T.iloc[1:]

def rate(df):
    # Getting coffee names 
    coffee_list = df.columns.to_list()


    # calculating gb-roast for each month of each line
    for col in range(1, df.shape[1]):
        if col % 2 == 1:  # Check if column index is odd
            prev_col = col - 1  # Get the previous column index
            df[col] = df[col] / df[prev_col]

    # calculating total green bean of each line
    even_column_sum = df.iloc[:, ::2].sum(axis=1)

    # Divide all even columns by their corresponding sum of even columns
    even_columns = df.iloc[:, ::2]
    df.iloc[:, ::2] = even_columns.div(even_column_sum, axis=0)

    # Add the sum as a new column
    df['each_line_sum'] = even_column_sum

    #calculating weighted rate of each line 
    rate_list =[]
    for i in range(0,len(df.index.to_list())):
        row_index = i
        row_values = df.iloc[row_index].tolist()
        # Deleting NAN
        cleaned_list = [x for x in row_values if pd.notna(x)]
        # Deleting sum value of each line in calculations
        value_list = cleaned_list[:-1]
        # portion of green bean list
        even_list = value_list[: :2]
        # rate from grean bean to sorted bean in each month of line
        odd_list = value_list[: :-2]

        # calculating weighted rate
        rate = 0
        for i in range(0,len(odd_list)):
            rate += (even_list[i] * odd_list[i])

        # adding to list 
        rate_list.append(rate)
        rate_list

    # adding to dataframe
    df["line_weighted_rate"] = rate_list

    # calculatng portion of eachline based on total grean bean
    df["each_line_portion"] = None
    total_kg = sum(df["each_line_sum"].to_list())

    for i in df.index.to_list():
        df.loc[i,"each_line_portion"] = df.loc[i,"each_line_sum"] / total_kg

    # calculatng weighted grean bean to roast rate 
    list1 = df["line_weighted_rate"].to_list()
    list2 = df["each_line_portion"].to_list()
    # Multiply corresponding elements from list1 and list2 and sum the results
    total_multiplication = 0
    for i in range(len(list1)):
        total_multiplication += list1[i] * list2[i]

    print((total_multiplication))
rate(df)


df = pd.read_excel("/Users/mohammad/Dropbox/XAV Finance (1)/گزارشات/1403/فروردین/نرخ تولید 1402.xlsx",sheet_name="roast-sort")[:-2].T.iloc[1:]
rate(df)

df = pd.read_excel("/Users/mohammad/Dropbox/XAV Finance (1)/گزارشات/1403/فروردین/نرخ تولید 1402.xlsx",sheet_name="sort-manualsort")[:-2].T.iloc[1:]
rate(df)
```
