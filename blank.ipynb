```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def distribution(data, title, bins=30, color='blue', save_path=None):
    """
    Create and display a histogram and density plot of the given data,
    including lines for the mean and standard deviation.

    Parameters:
    - data: List of numerical values.
    - title: Title for the plots.
    - bins: Number of bins for the histogram (default is 30).
    - color: Color for the plots (default is 'blue').
    - save_path: If provided, save the plots to this file path (default is None).
    """

    # Check if data is a list of numbers
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError("Data should be a list of numerical values.")

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)
    print(mean)

    # Create a histogram
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=bins, alpha=0.7, color=color, edgecolor='black')
    mean_line = plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
    std_dev_plus_line = plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1)
    std_dev_minus_line = plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1)
    plt.title(f'Histogram of {title}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend([mean_line, std_dev_plus_line, std_dev_minus_line], ['Mean', '+1 Std Dev', '-1 Std Dev'])
    if save_path:
        plt.savefig(f"{save_path}_histogram.png")
    plt.show()

    # Create a density plot
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data, shade=True, color=color)
    mean_line = plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
    std_dev_plus_line = plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1)
    std_dev_minus_line = plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1)
    plt.title(f'Density Plot of {title}')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend([mean_line, std_dev_plus_line, std_dev_minus_line], ['Mean', '+1 Std Dev', '-1 Std Dev'])
    if save_path:
        plt.savefig(f"{save_path}_density.png")
    plt.show()

    # Create both plots together
    plt.figure(figsize=(10, 6))
    sns.histplot(data, bins=bins, kde=True, color=color, edgecolor='black')
    mean_line = plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
    std_dev_plus_line = plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1)
    std_dev_minus_line = plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1)
    plt.title(f'Histogram and Density Plot of {title}')
    plt.xlabel('Value')
    plt.ylabel('Frequency / Density')
    plt.legend([mean_line, std_dev_plus_line, std_dev_minus_line], ['Mean', '+1 Std Dev', '-1 Std Dev'])
    if save_path:
        plt.savefig(f"{save_path}_combined.png")
    plt.show()

# Example usage
# distribution([1, 2, 3, 4, 5], "Sample Data", bins=5, color='green', save_path='output_plot')

```
