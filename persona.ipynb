```python
import pandas as pd


df = pd.read_excel("پرسش‌نامه ژاو.xlsx")
print(df.columns)
df["activity"].value_counts()
```

## coffee jobs

```python
coffee_job = df["coffee jobs"]
coffee_job = coffee_job.dropna()
coffee_job_dict = coffee_job.to_dict()
coffee_job_dict
```

```python
# counting every staff of coffee jobs
barista = []
manager = []
owner = []
staff = []

for i in coffee_job.index.tolist():
    job = coffee_job_dict[i]
    
    if job == "باریستا":
        barista .append(i)
    elif job == "مدیر کافه یا رستوران":
        manager.append(i)
    elif job == "صاحب کافه یا رستوران":
        owner.append(i)
    elif job == "توسعه و تحقیق در صنعت قهوه":
        staff.append(i)
    elif job == "فعالیت\u200cهای مربوط به برشته\u200cکاری قهوه":
        staff.append(i)
    elif job == "آشپز":
        staff.append(i)
    elif job == "بار تندر":
        barista.append(i)
```

## Non Coffee Jobs

```python
non_coffee = df["activity"]
non_coffee = non_coffee.dropna()
non_coffee_dict = non_coffee.to_dict()
non_coffee_dict
```

```python
# counting every non coffee jobs
coffee = []
employee = []
artist = []
engineer = []
medical = []
it = []

for i in non_coffee.index.tolist():
    job = non_coffee_dict[i]
    
    if job == "café":
        coffee.append(i)
    elif job == "employee":
        employee.append(i)
    elif job == "art":
        artist.append(i)
    elif job == "engineer":
        engineer.append(i)
    elif job == "medical":
        medical.append(i)
    elif job == "IT":
        it.append(i)
    
    
```

## cafe staff datframes

```python
barista_df = df[df.index.isin(barista)]
manager_df = df[df.index.isin(manager)]
owner_df = df[df.index.isin(owner)]
staff_df = df[df.index.isin(staff)]

```

## non coffee jobs Dataframes

```python
employee_df = df[df.index.isin(employee)]
artist_df = df[df.index.isin(artist)]
engineer_df = df[df.index.isin(engineer)]
medical_df = df[df.index.isin(medical)]
it_df = df[df.index.isin(it)]
```

## persona Function

```python
def persona(df):
    # sex
    men = pd.DataFrame(df["sex"]).value_counts()["مرد"]
    women = pd.DataFrame(df["sex"]).value_counts()["زن"]
    print(f"men --> {(men/ (men+women)) * 100 }%")
    print(f"women --> {(women/ (men+women)) *100} % \n \n")
    
    # age
    fourty_one = df["age"].value_counts()["۳۶ تا ۴۵ سال"]
    try:
        thirty_one = df["age"].value_counts()["۲۶ تا ۳۵ سال"]
        try:
            twenty_one = df["age"].value_counts()["۱۸ تا ۲۵ سال"]
            avg = ((twenty_one * 21) + (thirty_one *31) + (fourty_one*41)) / (twenty_one + thirty_one + fourty_one )
        except:
            avg = ((thirty_one *31) + (fourty_one*41)) / (thirty_one + fourty_one )
        
    except:
        avg = (fourty_one*41) / fourty_one
        

        
    print(f"the average of age = {avg} \n \n")
    
    # place
    first_place = df["living place"].value_counts().index.tolist()[0]
    second_place = df["living place"].value_counts().index.tolist()[1]
    third_place = df["living place"].value_counts().index.tolist()[2]
    print(f"the place of living --> {first_place} , {second_place} , {third_place} \n \n")
    
    # revenue
    m_31 = df["revenue"].value_counts()["بالای ۲۵ میلیون تومان"]
    try:
        m_21 = df["revenue"].value_counts()["۱۶ تا ۲۵ میلیون تومان"]
        try:
            m_13 = df["revenue"].value_counts()["۱۱ تا ۱۵ میلیون تومان"]
            try:
                m_8 = df["revenue"].value_counts()["۶ تا ۱۰ میلیون تومان"]
                try:
                    m_3 = df["revenue"].value_counts()["کمتر از ۵ میلیون تومان"] 
                    avg = ( (m_8 * 8) + (m_13 * 13) + (m_3 * 3) + (m_21 * 21) + (m_31 * 31) ) / (m_8 + m_13 + m_3 + m_21 + m_31)
                except:
                    avg = ( (m_8 * 8) + (m_13 * 13)  + (m_21 * 21) + (m_31 * 31) ) / (m_8 + m_13  + m_21 + m_31)
            except:
                avg = (  (m_13 * 13)  + (m_21 * 21) + (m_31 * 31) ) / ( m_13  + m_21 + m_31)

        except:
            avg = ( (m_21 * 21) + (m_31 * 31) ) / ( m_21 + m_31)
    except:
        avg = ( (m_31 * 31) ) / (m_31)
                    
        
    print(f"the average revenue = {avg} \n \n")
    
    
    # skills
    espresso = ((df["ماشین اسپرسو"].value_counts())[0] / len(df))*100
    v60 = ((df["وی۶۰، کمکس، کالیتا، اوریگامی (دمی چکانه‌ای دستی)"].value_counts())[0] / len(df))*100
    psiphon = ((df["ائروپرس، سایفون و ... (دمی با استفاده از خلأ)"].value_counts())[0] / len(df))*100
    french = ((df["فرنچ‌پرس، کلور و ... (دمی بر پایه همجواری)"].value_counts())[0] / len(df))*100
    mocha = ((df["موکاپات، جزوه و ... (دستگاه‌های سنتی)"].value_counts())[0] / len(df))*100


    print(f"espresso machine usage percent {espresso}%")
    print(f"v60,chemex,origami,kalita usage percent {v60}%")
    print(f"psiphon aeropress usage percent {psiphon}%")
    print(f"frenchpress usage percent {french}%")
    print(f"traditional gadgets usage percent {mocha}% \n \n")    
    
    # entertainment
    book = ((df["book"].value_counts())[0] / len(df))*100
    podcast = ((df["podcast"].value_counts())[0] / len(df))*100
    movies = ((df["movies"].value_counts())[0] / len(df))*100
    art = ((df["art"].value_counts())[0] / len(df))*100
    nature = ((df["nature"].value_counts())[0] / len(df))*100
    walking = ((df["walking/cycling"].value_counts())[0] / len(df))*100
    cafe = ((df["café/restaurant"].value_counts())[0] / len(df))*100
    print(f"reading book --> {book}%")
    print(f"podcast --> {podcast}%")
    print(f"movies --> {movies}%")
    print(f"art --> {art}%")
    print(f"nature --> {nature}%")
    print(f"walking/ cycling --> {walking}%")
    print(f"cafe/restaurant --> {cafe}%")
    try:
        cinema = ((df["cinema, theatre"].value_counts())[0] / len(barista_df))*100
        print(f"cinema/theatre --> {cinema}%\n \n")
    except:
        pass

    
    
    # media
    insta = ((df["how to follow"].value_counts())[0] / len(df))*100
    website = ((df["how to follow"].value_counts())[1] / len(df))*100

    print(f"insta --> {insta}%")
    print(f"website --> {website}% \n \n ")
    
    # 
    
    # source
    article = ((df["learning about coffee"].value_counts())[0] / len(df))*100
    video = ((df["learning about coffee"].value_counts())[1] / len(df))*100
    books = ((df["learning about coffee"].value_counts())[2] / len(df))*100
    print(f"article --> {article}%")
    print(f"video --> {video}%")
    print(f"books & magazine --> {books}%")
    try:
        other = ((df["learning about coffee"].value_counts())[3] / len(df))*100
        print(f"others --> {other}%\n \n ")  
    except:
        pass
```
