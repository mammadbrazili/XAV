```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.palettes import Paired
from bokeh.models import ColumnDataSource
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Spectral6
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import LinearColorMapper, ColorBar, LabelSet

from bokeh.models import LinearColorMapper
from bokeh.plotting import figure, show
from bokeh.sampledata.iris import flowers
import bokeh
```

```python

tehran = pd.read_excel("/Users/mohammad/Downloads/نرخ رشد و کاهش مشتریان دی و بهمن ماه.xlsx", sheet_name="تهران")
tehran = tehran.dropna()
growth_rates_1 = tehran["نرخ رشد یا کاهش در دو ماه %"].to_list()
tehran =  tehran
cafe_list = tehran["مشتري"].to_list()
day_buy =  tehran["مقدار اصلی (کیلوگرم) / دی ماه"].to_list()
growth_rates = tehran["نرخ رشد یا کاهش در دو ماه %"].to_list()
growth_rates = [round(x, 2) for x in growth_rates]

mapper = linear_cmap(field_name='growth_rates', palette=Spectral6[: : -1], low=min(growth_rates_1), high=max(growth_rates_1))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(customers=cafe_list, kg_bought=day_buy, growth_rates=growth_rates))

# Create the figure and plot the data
p = figure(x_range=cafe_list, height=800, width = 1450, title="Customer KG Bought", toolbar_location=None, tools="")

# Plot the data using circles and the color mapper
p.circle(x='customers', y='kg_bought', size=15, source=source, line_color=mapper, fill_color=mapper)
labels = LabelSet(x='customers', y='kg_bought', text='growth_rates', level='glyph',
              x_offset=5, y_offset=5, source=source, text_font_size='7pt')
p.add_layout(labels)
# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')
p.xaxis.major_label_orientation = 45
p.yaxis.axis_label = "kg" # Add the y-axis title



# Show the plot
show(p)






```
